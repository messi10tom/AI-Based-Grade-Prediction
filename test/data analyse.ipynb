```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = pd.read_csv("../data/student/student-mat.csv", sep=";")
data_1.head()
```

```python
data_1.describe()
```

```python
# family size (binary: "LE3" - less or equal to 3 or "GT3" - greater than 3)

famsize_values = data_1['famsize'].value_counts()

famsize_values.plot(kind='bar')
print(f"Percentage of GT3: {(famsize_values['GT3'] / sum(famsize_values)) * 100:.2f}")
print(f"Percentage of LE3: {(famsize_values['LE3'] / sum(famsize_values)) * 100:.2f}")
```

```python
columns_to_plot = [
    'school', 'sex', 'address', 'famsize', 'Pstatus', 
    'schoolsup', 'famsup', 'paid', 'activities', 
    'nursery', 'higher', 'internet', 'romantic'
]

# Plot the distribution of all columns in a single graph
fig, ax = plt.subplots(figsize=(15, 10))
data_1[columns_to_plot].apply(pd.Series.value_counts).T.plot(kind='bar', stacked=True, ax=ax)
plt.title('Distribution of Categorical Variables')
plt.ylabel('Count')
plt.xlabel('Variables')
plt.xticks(rotation=45)
plt.legend(title='Values', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
