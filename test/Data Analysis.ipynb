```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_1 = pd.read_csv("../data/student/student-mat.csv", sep=";")
data_1
```

```python
data_1.describe()
```

```python
columns_to_plot = [
    'school', 'sex', 'address', 'famsize', 'Pstatus', 
    'schoolsup', 'famsup', 'paid', 'activities', 
    'nursery', 'higher', 'internet', 'romantic'
]

# Plot the distribution of all columns in a single graph
fig, ax = plt.subplots(figsize=(15, 10))
data_1[columns_to_plot].apply(pd.Series.value_counts).T.plot(kind='bar', stacked=True, ax=ax)
plt.title('Distribution of Categorical Variables')
plt.ylabel('Count')
plt.xlabel('Variables')
plt.xticks(rotation=45)
plt.legend(title='Values', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Data Analysis and Visualization

In this analysis, we are exploring the distribution of various categorical variables in the student dataset. The graph below shows the distribution of these variables:

<img src="The_data_distribution.png" alt="Distribution of Categorical Variables" width="33%">


#### Observations:

1. **School**: The dataset contains data from only two schools, 'GP' and 'MS'. More than 80% of the students belong to the 'GP' school. This imbalance means that the 'school' variable may not provide significant insights for our analysis, as it lacks diversity.

2. **Sex**: The distribution of 'sex' is relatively balanced, with a slight majority of female students. This variable can be useful for gender-based analysis.

3. **Address**: The 'address' variable indicates whether a student lives in an urban ('U') or rural ('R') area. This variable has a fairly balanced distribution and can be useful for understanding the impact of living environment on student performance.

4. **Family Size (famsize)**: The 'famsize' variable shows whether a student comes from a family with more than 3 members (GT3) or less (LE3). This variable is also fairly balanced and can provide insights into the impact of family size on student performance.

5. **Parental Status (Pstatus)**: This variable indicates whether the student's parents live together ('T') or apart ('A'). The majority of students have parents who live together, but there is still a significant portion with parents living apart, making this variable useful for analysis.

6. **Support Programs**: Variables like 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', and 'romantic' indicate whether students receive various forms of support or participate in activities. These variables have diverse distributions and can provide valuable insights into the impact of these factors on student performance.

7. **Other Variables**: Variables such as 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime', 'failures', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2', and 'G3' provide additional context about the students' backgrounds and behaviors. These variables have varying distributions and can be useful for a comprehensive analysis.

In conclusion, while some variables like 'school' may not be very informative due to their lack of diversity, other variables provide a wealth of information that can be used to understand the factors influencing student performance. By focusing on the more diverse and balanced variables, we can gain deeper insights into the dataset.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot Age vs G1
axes[0].scatter(data_1['age'], data_1['G1'], alpha=0.25, color='black')
axes[0].set_title('Age vs G1')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('G1')

# Plot Age vs G2
axes[1].scatter(data_1['age'], data_1['G2'], alpha=0.25, color='black')
axes[1].set_title('Age vs G2')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('G2')

# Plot Age vs G3
axes[2].scatter(data_1['age'], data_1['G3'], alpha=0.25, color='black')
axes[2].set_title('Age vs G3')
axes[2].set_xlabel('Age')
axes[2].set_ylabel('G3')

plt.tight_layout()
plt.show()
```

```python
# Setting up the plot
plt.figure(figsize=(12, 8))

# Bar plot for Medu and Fedu
plt.subplot(2, 1, 1)
sns.barplot(data=data_1, x='Medu', y='G3', hue='Fedu', errorbar=None, palette='coolwarm')
plt.title('Final Grade (G3) by Parents\' Education Levels')
plt.xlabel('Mother\'s Education Level (Medu)')
plt.ylabel('Final Grade (G3)')
plt.legend(title='Father\'s Education Level (Fedu)', bbox_to_anchor=(1.05, 1), loc='upper left')

# Scatter plot for grades
plt.subplot(2, 1, 2)
scatter = sns.scatterplot(data=data_1, x='G1', y='G2', size='G3', hue='G3', sizes=(20, 200), palette='viridis')
plt.title('Grade Comparison and Final Grade Influence')
plt.xlabel('First Period Grade (G1)')
plt.ylabel('Second Period Grade (G2)')
scatter.legend(title='Final Grade (G3)', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# Create a combined column for Mjob and Fjob for easier plotting
df_mjob_fjob = pd.melt(data_1, value_vars=['Mjob', 'Fjob'], var_name='Job Type', value_name='Job Category')

# Create the count plot
plt.figure(figsize=(10, 6))
sns.countplot(data=df_mjob_fjob, x='Job Category', hue='Job Type', palette='Set2')

# Add title and labels
plt.title('Distribution of Mother\'s and Father\'s Jobs')
plt.xlabel('Job Categories')
plt.ylabel('Count')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Create the count plot
plt.figure(figsize=(10, 6))
sns.countplot(data=data_1, x='reason')

# Add title and labels
plt.title('Distribution of Reason to Choose School')
plt.xlabel('Reasons')
plt.ylabel('Count')

# Display the plot
plt.tight_layout()
plt.show()
```
